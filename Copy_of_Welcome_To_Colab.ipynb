```
import torchvision.transforms as transforms
import random
from torch.utils.data import Dataset
import cv2
import torch

class ChangeDetectionDataset(Dataset):
    def __init__(self, img1_paths, img2_paths, label1_paths, label2_paths, augment=False):
        self.img1_paths = img1_paths
        self.img2_paths = img2_paths
        self.label1_paths = label1_paths
        self.label2_paths = label2_paths
        self.augment = augment

        # Preprocessing Steps:
        # 1. Resize images and labels to 256x256 for uniform input size
        # 2. Convert images to tensor and normalize pixel values between 0 and 1
        # 3. For labels, convert to tensor (grayscale)
        # 4. Data augmentation (optional): random horizontal flip and rotations for robustness

        self.img_transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((256,256)),
            transforms.ToTensor(),  # scales pixel values 0-1
        ])

        self.mask_transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((256,256)),
            transforms.ToTensor(),
        ])

    def __len__(self):
        return len(self.img1_paths)

    def __getitem__(self, idx):
        img1 = cv2.imread(self.img1_paths[idx])
        img2 = cv2.imread(self.img2_paths[idx])
        label1 = cv2.imread(self.label1_paths[idx], 0)
        label2 = cv2.imread(self.label2_paths[idx], 0)

        if self.augment:
            # Apply augmentations here
            if random.random() > 0.5:
                img1 = cv2.flip(img1, 1)
                img2 = cv2.flip(img2, 1)
                label1 = cv2.flip(label1, 1)
                label2 = cv2.flip(label2, 1)

            angle = random.choice([0,90,180,270])
            if angle != 0:
                rotate_code = {90:cv2.ROTATE_90_CLOCKWISE, 180:cv2.ROTATE_180, 270:cv2.ROTATE_90_COUNTERCLOCKWISE}[angle]
                img1 = cv2.rotate(img1, rotate_code)
                img2 = cv2.rotate(img2, rotate_code)
                label1 = cv2.rotate(label1, rotate_code)
                label2 = cv2.rotate(label2, rotate_code)

        # Apply transforms after augmentation
        img1 = self.img_transform(img1)
        img2 = self.img_transform(img2)
        label1 = self.mask_transform(label1)
        label2 = self.mask_transform(label2)

        change_mask = (label1 != label2).float()  # binary change mask

        return img1, img2, change_mask


# Created dataset with preprocessing included
train_dataset = ChangeDetectionDataset(train_img1_paths, train_img2_paths, train_label1_paths, train_label2_paths, augment=True)
test_dataset = ChangeDetectionDataset(test_img1_paths, test_img2_paths, test_label1_paths, test_label2_paths, augment=False)


# Calling the file consistency check function again as before to verify everything is correct
import os

check_dataset_consistency(train_img1_paths, train_img2_paths, train_label1_paths, train_label2_paths, "Train Dataset")
check_dataset_consistency(test_img1_paths, test_img2_paths, test_label1_paths, test_label2_paths, "Test Dataset")
```

```
#Dataset Loader with Preprocessing + Augmentation
import torchvision.transforms as transforms
import random
from torch.utils.data import Dataset
import cv2
import torch

class ChangeDetectionDataset(Dataset):
    def __init__(self, img1_paths, img2_paths, label1_paths, label2_paths, augment=False):
        self.img1_paths = img1_paths
        self.img2_paths = img2_paths
        self.label1_paths = label1_paths
        self.label2_paths = label2_paths
        self.augment = augment

        # Image transform (resize -> tensor -> normalize)
        self.img_transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((256, 256)),
            transforms.ToTensor(),                # scales [0,255] → [0,1]
            transforms.Normalize([0.5]*3, [0.5]*3)  # optional: scale → [-1,1]
        ])

        # Mask transform (grayscale resize -> tensor)
        self.mask_transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((256, 256)),
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.img1_paths)

    def __getitem__(self, idx):
        # Load images
        img1 = cv2.imread(self.img1_paths[idx])
        img2 = cv2.imread(self.img2_paths[idx])
        label1 = cv2.imread(self.label1_paths[idx], cv2.IMREAD_GRAYSCALE)
        label2 = cv2.imread(self.label2_paths[idx], cv2.IMREAD_GRAYSCALE)

        # --- Data Augmentation ---
        if self.augment:
            # Horizontal flip
            if random.random() > 0.5:
                img1 = cv2.flip(img1, 1)
                img2 = cv2.flip(img2, 1)
                label1 = cv2.flip(label1, 1)
                label2 = cv2.flip(label2, 1)

            # Rotation (0,90,180,270)
            angle = random.choice([0, 90, 180, 270])
            if angle != 0:
                rotate_code = {
                    90: cv2.ROTATE_90_CLOCKWISE,
                    180: cv2.ROTATE_180,
                    270: cv2.ROTATE_90_COUNTERCLOCKWISE
                }[angle]
                img1 = cv2.rotate(img1, rotate_code)
                img2 = cv2.rotate(img2, rotate_code)
                label1 = cv2.rotate(label1, rotate_code)
                label2 = cv2.rotate(label2, rotate_code)

        # --- Transform to tensor & resize ---
        img1 = self.img_transform(img1)
        img2 = self.img_transform(img2)
        label1 = self.mask_transform(label1)
        label2 = self.mask_transform(label2)

        # Binary change mask
        change_mask = (label1 != label2).float()

        return img1, img2, change_mask
```

```
#🔹 Dataset & Loader'
# Train dataset with augmentation
train_dataset = ChangeDetectionDataset(
    train_img1_paths, train_img2_paths,
    train_label1_paths, train_label2_paths,
    augment=True
)

# Test dataset without augmentation
test_dataset = ChangeDetectionDataset(
    test_img1_paths, test_img2_paths,
    test_label1_paths, test_label2_paths,
    augment=False
)

from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=4)
```

**Step 2: Siamese U-Net Model (Change Detection জন্য)**

```
import torch
import torch.nn as nn
import torch.nn.functional as F

# --- U-Net Encoder Block ---
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ConvBlock, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.conv(x)

# --- Siamese U-Net ---
class SiameseUNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super(SiameseUNet, self).__init__()

        # Encoder
        self.enc1 = ConvBlock(in_channels, 64)
        self.enc2 = ConvBlock(64, 128)
        self.enc3 = ConvBlock(128, 256)
        self.enc4 = ConvBlock(256, 512)

        self.pool = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = ConvBlock(512, 1024)

        # Decoder
        self.up4 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.dec4 = ConvBlock(1024, 512)

        self.up3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.dec3 = ConvBlock(512, 256)

        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.dec2 = ConvBlock(256, 128)

        self.up1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.dec1 = ConvBlock(128, 64)

        # Output layer
        self.final = nn.Conv2d(64, out_channels, kernel_size=1)

    def encode(self, x):
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))
        e3 = self.enc3(self.pool(e2))
        e4 = self.enc4(self.pool(e3))
        b = self.bottleneck(self.pool(e4))
        return e1, e2, e3, e4, b

    def forward(self, img1, img2):
        # Encode both images
        e1a, e2a, e3a, e4a, ba = self.encode(img1)
        e1b, e2b, e3b, e4b, bb = self.encode(img2)

        # Feature difference
        e1 = torch.abs(e1a - e1b)
        e2 = torch.abs(e2a - e2b)
        e3 = torch.abs(e3a - e3b)
        e4 = torch.abs(e4a - e4b)
        b  = torch.abs(ba - bb)

        # Decoder path
        d4 = self.up4(b)
        d4 = torch.cat([d4, e4], dim=1)
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        d3 = torch.cat([d3, e3], dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.dec1(d1)

        out = self.final(d1)
        return torch.sigmoid(out)   # probability map
```

**Step 3: Training Loop**

```
from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Model
model = SiameseUNet(in_channels=3, out_channels=1).to(device)

# Loss function (binary mask → BCE loss)
criterion = nn.BCELoss()

# Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# Training
num_epochs = 20

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for img1, img2, mask in train_loader:
        img1, img2, mask = img1.to(device), img2.to(device), mask.to(device)

        optimizer.zero_grad()
        outputs = model(img1, img2)

        loss = criterion(outputs, mask)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")
```

**🔹 Step 4: Evaluation on Test Set**

```
from sklearn.metrics import accuracy_score

model.eval()
all_preds, all_labels = [], []

with torch.no_grad():
    for img1, img2, mask in test_loader:
        img1, img2, mask = img1.to(device), img2.to(device), mask.to(device)

        outputs = model(img1, img2)
        preds = (outputs > 0.5).float()

        all_preds.extend(preds.cpu().numpy().flatten())
        all_labels.extend(mask.cpu().numpy().flatten())

acc = accuracy_score(all_labels, all_preds)
print("Test Accuracy:", acc)
```
