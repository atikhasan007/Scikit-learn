```python
import pandas as pd
df = pd.read_csv('/content/homeprices.csv')
df.head()
```

```python
df['town'].unique()
```

```python
dummies = pd.get_dummies(df.town).astype(int)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 1)
```

```python
merged.head()
```

```python
final = merged.drop(['town'],axis=1)
```

```python
final = merged.drop(['town'], axis='columns')
final = final[['monroe township', 'robinsville', 'west windsor', 'area', 'price']]
final
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['town']]).toarray()
encoded
```

```python
encoded_df = pd.DataFrame(data=encoded, columns=encoder.enget_feature_names_out())
```
