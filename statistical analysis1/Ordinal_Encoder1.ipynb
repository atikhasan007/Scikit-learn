```python
#Ordinal Encoding ঠিক কাজ করে যখন
#category-গুলোর মধ্যে order/priority থাকে (Ordinal Data),
# কিন্তু Nominal Data-তে ভুল signal দেয়।
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/shop data.csv')
df.head()
```

```python
#ordinal encoding
df['income'].unique()
```

```python
income_order = ['low','medium','high']
```

```python
from sklearn.preprocessing import OrdinalEncoder

enco = OrdinalEncoder(categories=[income_order])
enco.fit(df[['income']])
```

```python
encode_data =enco.transform(df[['income']])
encode_data
```

```python
final = pd.DataFrame(data= encode_data, columns=['income'])
final
```

```python
df['income'] = final
df
```
