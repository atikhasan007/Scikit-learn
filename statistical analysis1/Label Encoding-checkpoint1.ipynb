```python
#label encoding is  easy and simple  and also good fit
#ordinal data but not good fit nominal data
import pandas as pd
df = pd.read_csv('/content/shop data.csv')
```

```python
x = df.drop('buys',axis=1)
```

```python
y = df['buys']
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_data = x.apply(le.fit_transform)
encoded_data
```
