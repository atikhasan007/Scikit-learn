```
import pandas as pd
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/History_of_Wikipedia"

# Custom header ‡¶¶‡¶ø‡ßü‡ßá ‡¶∞‡¶ø‡¶ï‡ßã‡ßü‡ßá‡¶∏‡ßç‡¶ü ‡¶™‡¶æ‡¶†‡¶æ‡¶ì
headers = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64)"}
response = requests.get(url, headers=headers)
response.raise_for_status()

# ‡¶∏‡¶¨ ‡¶ü‡ßá‡¶¨‡¶ø‡¶≤ ‡¶™‡ßú‡ßã
df_list = pd.read_html(StringIO(response.text))

# ‡¶ï‡ßü‡¶ü‡¶æ ‡¶ü‡ßá‡¶¨‡¶ø‡¶≤ ‡¶™‡ßá‡¶≤‡¶æ‡¶Æ ‡¶¶‡ßá‡¶ñ‡¶ø
print(f"Found {len(df_list)} tables\n")

# ‡¶™‡ßç‡¶∞‡¶•‡¶Æ ‡¶ü‡ßá‡¶¨‡¶ø‡¶≤‡¶ü‡¶æ DataFrame ‡¶π‡¶ø‡¶∏‡ßá‡¶¨‡ßá ‡¶®‡¶ø‡¶á
df = df_list[0]

# DataFrame ‡¶¶‡ßá‡¶ñ‡¶æ
print(df.head())
```
