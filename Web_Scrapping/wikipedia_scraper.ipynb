```
import pandas as pd
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/History_of_Wikipedia"

# Custom header দিয়ে রিকোয়েস্ট পাঠাও
headers = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64)"}
response = requests.get(url, headers=headers)
response.raise_for_status()

# সব টেবিল পড়ো
df_list = pd.read_html(StringIO(response.text))

# কয়টা টেবিল পেলাম দেখি
print(f"Found {len(df_list)} tables\n")

# প্রথম টেবিলটা DataFrame হিসেবে নিই
df = df_list[0]

# DataFrame দেখা
print(df.head())
```
